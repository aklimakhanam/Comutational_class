```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for Euler's method
def Euler(N0, lambda_dc, ti, tf, dt):
    
    #initialization of variables and lists
    times = [ti]
    quantities = [N0]
    activity = []
    dNdt = []
    hlife = []
    t = ti
    N = N0
    
    #Loop for numerical calculation
    while t < tf:
        dN_dt = -lambda_dc * N
        N += dN_dt * dt
        t += dt
        times.append(t)
        quantities.append(N)
    
    #Loop for verifying activity
    for i in range(0, len(quantities)):
        if i==0:
            continue
        else:
            rate = (quantities[i] - quantities[i-1]) / dt
            test_rate = -lambda_dc * quantities[i]
            activity.append(rate)
            dNdt.append(test_rate)
        
    #Loop for verifying half life
    for i in range(0, len(quantities)):
        if quantities[i]>=(0.5*N0 - 0.01*N0) and quantities[i]<= (0.5*N0 + 0.01*N0):
            hlife.append(times[i])
            #print("Probable half-lifes: ", hlife)
    print(hlife)
    return times, quantities, activity, dNdt


# Define the function for RK4 method
def RK4(N0, lambda_dc, ti, tf, dt):
    
    #initialization of variables and lists
    times = [ti]
    quantities = [N0]
    activity = []
    dNdt = []
    hlife = []
    t = ti
    N = N0
    
    def dN_dt(N, t):
        return -lambda_dc * N

    while t < t_end:
        k1 = dt * dN_dt(N, t)
        k2 = dt * dN_dt(N + 0.5 * k1, t + 0.5 * dt)
        k3 = dt * dN_dt(N + 0.5 * k2, t + 0.5 * dt)
        k4 = dt * dN_dt(N + k3, t + dt)
        
        N += (k1 + 2 * k2 + 2 * k3 + k4) / 6
        t += dt
        times.append(t)
        quantities.append(N)
    
    return times, quantities

# Parameters
N0 = 1000  # Initial quantity
lambda_dc = 0.1  # Decay constant
ti = 0  # Initial time
tf = 10  # End time
dt = 0.1  # Time step

# Solve using Euler's method
times_euler, quantities_euler, activity, dNdt = Euler(N0, lambda_dc, ti, tf, dt)

# Solve using RK4 method
times_rk4, quantities_rk4 = rk4_method(N0, lambda_dc, ti, tf, dt)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(times_euler, quantities_euler, label="Euler's Method")
plt.plot(times_rk4, quantities_rk4, label="RK4 Method")
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.legend()
plt.title('Radioactive Decay using Euler\'s Method and RK4 Method')
plt.grid(True)
plt.show()
```
